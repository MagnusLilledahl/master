# Scanning template
The scanning template holds all information about a scan. This notebook will explore how to read, parse and edit such templates.

```python
# metadata holds files from LASAF data exporter
!ls metadata
```

```python
def xmlprint(filename):
    "Pretty prints XML in filename."
    import xml.dom.minidom

    dom = xml.dom.minidom.parse(filename) # or xml.dom.minidom.parseString(xml_string)
    print(dom.toprettyxml())
```

```python
template = "metadata/{ScanningTemplate}test_CAM.xml"
xmlprint(template)
```

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse(template)
```

```python
root = tree.getroot()
```

```python
print(root.tag, root.attrib)
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
from pprint import pprint
print(root[1][0])
pprint(root[1][0].attrib)
```

```python
from bunch import Bunch # for easy notation
```

```python
%qtconsole
```

```python
tmp = {}
x_fields = root.find("ScanningTemplate").find("CountOfScanFieldsX")
y_fields = root.find("ScanningTemplate").find("CountOfScanFieldsY")
for field in root.iter('ScanFieldData'):
    attrib = Bunch(field.attrib)
    well_name = attrib.LabelX + attrib.LabelY
    msg = ""
    x = float(field.find('FieldXCoordinate').text)
    y = float(field.find('FieldYCoordinate').text)
    if well_name not in tmp.keys():
        tmp[well_name] = True
        msg = "Well - " + well_name + "\n"
    msg += "Field: {FieldX},{FieldY} - ".format(**attrib)
    msg += "Pos: {},{}".format(int(x*1e6),int(y*1e6))
    print(msg)
```
