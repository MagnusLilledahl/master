````python
import xml.etree.ElementTree as ET

def offset_well(filename, u, v, offset_x=0, offset_y=0, offset_z=0):
    "Offset a given well in a etree xml."
    
    tree = ET.parse(filename)
    root = tree.getroot()

    # every ScanFieldData with attr WellX and WellY set to u,v
    # https://docs.python.org/3/library/xml.etree.elementtree.html#supported-xpath-syntax
    fields = root.findall('.//ScanFieldData[@WellX="{}"][@WellY="{}"]'.format(u,v))
    # set field values
    for field in fields:
        X = field.find('FieldXCoordinate')
        Y = field.find('FieldYCoordinate')
        Z = field.find('FieldZCoordinate')

        x = float(X.text) + offset_x
        print(x)
        y = float(Y.text) + offset_y
        z = float(Z.text) + offset_z

        X.text = "%.19f" % x
        Y.text = "%.19f" % y
        Z.text = "%.19f" % z

    # set well start value
    well = root.find('.//ScanWellData[@WellX="{}"][@WellY="{}"]'.format(u,v))
    
    x_start = float(well.attrib['FieldXStartCoordinate']) + offset_x
    y_start = float(well.attrib['FieldYStartCoordinate']) + offset_y
    z_start = float(well.attrib['FieldZCoordinate']) + offset_z
    
    well.attrib['FieldXStartCoordinate'] = "%.19f" % x_start
    well.attrib['FieldYStartCoordinate'] = "%.19f" % y_start
    well.attrib['FieldZCoordinate'] = "%.19f" % z_start

    tree.write(filename)
    
    return tree
````

````python
def prettyxml(filename):
    "Returns xml as pretty string."
    import xml.dom.minidom
    import re

    dom = xml.dom.minidom.parse(filename) # or xml.dom.minidom.parseString(xml_string)
    output = dom.toprettyxml().splitlines()
    def not_whitespace(string):
        "Return false if string is considered whitespace."
        patterns = [
            re.compile(r'^\t+$'), #only tabs
            re.compile(r'^ +$'),  #only spaces
            re.compile(r'^$') ]   #empty
        for pattern in patterns:
            if pattern.match(string) is not None:
                return False
        return True
    output = [line for line in output if not_whitespace(line)]
    return '\n'.join(output)
````
