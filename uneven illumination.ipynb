```python
%matplotlib inline
from skimage import io
```

```python
img = io.imread('data/uneven_illumination.tif')[0,:,:] # red channel
```

```python
img.shape, img.dtype, img.max(), img.min(), img.max()-img.min()
```

```python
from skimage import img_as_ubyte
img = img_as_ubyte(img)
```

```python
io.imshow(img, vmin=img.min(), vmax=img.max())
```

```python
from skimage import filters, morphology
from matplotlib import pyplot as plt
```

```python
# remove noise with rank median (similar to histogram center)
selem = morphology.disk(5)
filtered = filters.rank.median(img, selem)
io.imshow(filtered, vmin=filtered.min(), vmax=filtered.max())
plt.colorbar();
```

```python
plt.contour(filtered)
plt.colorbar();
```
