Explore options for compressing images (without loosing TIFF tag information).

Implementation should be something like:

```
compress(images, delete_tiff=false):
  
    if compressed exists: do nothing, notify
    for every img:
        compress
        if delete_tiff: delete

decompress(images, delete_png=false):
  
    if decompressed exists: do nothing, notify
    for every img:
        decompress
        if delete_png: delete
```

````python
from PIL import Image
````

````python
import matrixscreener.experiment as matexp
````

````python
experiment = matexp.Experiment('data/experiment--row 0-2/')
````

````python
img = Image.open(experiment.images[0])
````

````python
print("\t".join(p for p in dir(img)))
````

````python
img.tag.as_dict()
````

````python
from PIL import TiffTags
````

````python
TiffTags.TAGS
````

````python
for tag in img.tag:
    print(TiffTags.TAGS[tag])
````

````python
for tag in img.tag:
    val = img.tag.as_dict()[tag]
    print(TiffTags.TAGS[tag] + "({})".format(type(val)), val)
````

````python
from PIL.PngImagePlugin import PngInfo
````

````python
tags = img.tag.as_dict()

info = PngInfo()
for tag in tags:
    name = TiffTags.TAGS[tag]
    value = tags[tag]
    if type(value) == tuple:
        value = "(" + ",".join(str(t) for t in value) + ")"
    info.add_text(name, value)
````

````python
info.chunks
````

````python
img.save('test.png', pnginfo = info)
````
